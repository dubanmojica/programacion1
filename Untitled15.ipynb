```python
def pedir_vector(mensaje):
    while True:
        try:
            entrada = input(mensaje).split()
            vector = [float(x) for x in entrada]
            return vector
        except ValueError:
            print("Error: Por favor, ingrese datos que sean numericos y que tengan la misma dimension.")

def suma_vectores(v1, v2):
    if len(v1) != len(v2):
        return "Error:Por favor, ingrese datos que sean numericos y que tengan la misma dimension."
    return [v1[i] + v2[i] for i in range(len(v1))]

def producto_punto(v1, v2):
    if len(v1) != len(v2):
        return "Error: Los vectores deben tener la misma dimensión."
    return sum(v1[i] * v2[i] for i in range(len(v1)))

def producto_hadamard(v1, v2):
    if len(v1) != len(v2):
        return "Error: Los vectores deben tener la misma dimensión."
    return [v1[i] * v2[i] for i in range(len(v1))]

def producto_cruz(v1, v2):
    if len(v1) == 3 and len(v2) == 3:
        x = v1[1]*v2[2] - v1[2]*v2[1]
        y = v1[2]*v2[0] - v1[0]*v2[2]
        z = v1[0]*v2[1] - v1[1]*v2[0]
        return [x, y, z]
    else:
        return "Error: El producto cruz solo está definido para vectores de dimensión 3."

def menu():
    while True:
        print("\n--- MENÚ DE OPERACIONES ---")
        print("1. Suma de vectores")
        print("2. Producto punto")
        print("3. Producto cruz (R3)")
        print("4. Producto de Hadamard")
        print("5. Salir")

        opcion = input("\nSeleccione una opción: ")

        if opcion == '5':
            print("Saliendo del programa...")
            break

        if opcion in ['1', '2', '3', '4']:
            vec1 = pedir_vector("Ingrese el vector 1 (separado por espacios): ")
            vec2 = pedir_vector("Ingrese el vector 2 (separado por espacios): ")

            if opcion == '1':
                print("Resultado Suma:", suma_vectores(vec1, vec2))
            elif opcion == '2':
                print("Resultado Producto Punto:", producto_punto(vec1, vec2))
            elif opcion == '3':
                print("Resultado Producto Cruz:", producto_cruz(vec1, vec2))
            elif opcion == '4':
                print("Resultado Hadamard:", producto_hadamard(vec1, vec2))
        else:
            print("Opción no válida. Intente de nuevo.")
menu()
```

```python
import numpy as np

def ingresar_matriz():
    while True:
        try:
            filas = int(input("Ingrese el número de filas: "))
            columnas = int(input("Ingrese el número de columnas: "))
            break
        except ValueError:
            print("Error: Ingrese un número entero válido.")

    matriz = []
    for i in range(filas):
        while True:
            try:
                entrada = input(f"Ingrese la fila {i+1} separada por espacios: ").split()
                fila = list(map(float, entrada))

                if len(fila) != columnas:
                    print(f"Error: debe ingresar exactamente {columnas} valores.")
                else:
                    matriz.append(fila)
                    break
            except ValueError:
                print("Error: Ingrese solo números.")

    return np.array(matriz)

def mostrar_menu():
    print("\n--- MENÚ DE OPERACIONES MATRICIALES ---")
    print("1. Suma de matrices")
    print("2. Multiplicación de matrices (Dot Product)")
    print("3. Producto de Hadamard (Elemento a elemento)")
    print("4. Producto de Kronecker")
    print("5. Salir")
    return input("Elija una opción: ")

def ejecutar():
    while True:
        opcion = mostrar_menu()

        if opcion == '5':
            print("Saliendo del programa...")
            break

        if opcion in ['1', '2', '3', '4']:
            print("\n--- Configuración de la Matriz A ---")
            A = ingresar_matriz()
            print("\n--- Configuración de la Matriz B ---")
            B = ingresar_matriz()

            try:
                if opcion == '1':
                    resultado = A + B
                    print("\nResultado de la Suma:\n", resultado)

                elif opcion == '2':
                    resultado = np.matmul(A, B)
                    print("\nResultado de la Multiplicación:\n", resultado)

                elif opcion == '3':
                    resultado = A * B
                    print("\nResultado del Producto de Hadamard:\n", resultado)

                elif opcion == '4':
                    resultado = np.kron(A, B)
                    print("\nResultado del Producto de Kronecker:\n", resultado)

            except ValueError as e:
                print(f"\nError Matemático: Las dimensiones de las matrices no son parejas.")
        else:
            print("Opción no válida. Intente de nuevo.")

if __name__ == "__main__":
    ejecutar()
```

```python
import numpy as np

def ingresar_matriz():
    while True:
        try:
            filas = int(input("Ingrese el número de filas: "))
            cols = int(input("Ingrese el número de columnas: "))
            break
        except ValueError:
            print("Error: Ingrese números enteros.")

    matriz = []
    for i in range(filas):
        while True:
            try:
                fila = list(map(float, input(f"Fila {i+1}: ").split()))
                if len(fila) != cols:
                    print(f"Error: ingrese {cols} valores.")
                else:
                    matriz.append(fila)
                    break
            except ValueError:
                print("Error: Ingrese solo números.")
    return np.array(matriz)

def obtener_cofactores(matriz):
    """Calcula la matriz de cofactores manualmente."""
    filas, cols = matriz.shape
    if filas != cols:
        raise ValueError("La matriz de cofactores solo se define para matrices cuadradas.")

    cofactores = np.zeros((filas, cols))
    for f in range(filas):
        for c in range(cols):
            submatriz = np.delete(np.delete(matriz, f, axis=0), c, axis=1)
            cofactores[f, c] = ((-1)**(f + c)) * np.linalg.det(submatriz)

    return cofactores

def ejecutar():
    while True:
        print("\n--- MENÚ DE TRANSFORMACIONES ---")
        print("1. Trasponer matriz")
        print("2. Matriz de cofactores")
        print("3. Vectorizar por filas")
        print("4. Salir")

        opcion = input("Elija una opción: ")

        if opcion == '4':
            break

        if opcion in ['1', '2', '3']:
            matriz = ingresar_matriz()

            try:
                if opcion == '1':
                    print("\nMatriz Traspuesta:\n", matriz.T)

                elif opcion == '2':
                    res = obtener_cofactores(matriz)
                    print("\nMatriz de Cofactores:\n", res)

                elif opcion == '3':
                    vector = matriz.flatten()
                    print("\nVectorización (vec(A)):\n", vector.reshape(-1, 1))

            except Exception as e:
                print(f"\nError: {e}")
        else:
            print("Opción no válida.")

if __name__ == "__main__":
    ejecutar()
```
