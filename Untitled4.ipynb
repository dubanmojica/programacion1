```python
print(set())

print(set([5, 2, 5, 1, 1.5]))
print(set((5, 2, 5, 1, 1.5)))
print(set("52511.5"))

conjunto = set([2, 3, 3, 4])
conjunto_2 = set([5, 3, 5, 6])
conjunto_3 = set([4, 2])

print(conjunto)
print(conjunto_2)
print(conjunto_3)

conjunto.add(1) # adiciona un elemento
print(conjunto)

conjunto.remove(1) # elimina un elemento
print(conjunto)

print(conjunto.intersection(conjunto_2)) #se utiliza el metodo intersecsion que envuelve los elementos de los dos conjuntos y saber la interseccion
print(conjunto_2.issubset(conjunto)) # se utiliza issubset para saber si los elemntos de un conjunto esta incluido en otro conjunto
print(conjunto_3.issubset(conjunto))
```

```python
lista = list((29, True, 3.1415, "El número de Avogadro sí que  es horrible"))

print(lista)
print(lista[3]) # Equivalente a lista[-1]
print([lista[1], lista[2]]) # Construyendo la lista manualmente en vez de usar slicing [1:3]

lista[2] = [3, 2, 1]
print(lista)

print(lista.__len__()) # Equivalente a la función len(lista)
```

```python
lista_nueva = [1, 2, 3, 4, 5]

lista_nueva += [3] # El operador += equivale a .append()

print(lista_nueva)

# Sumamos 1 por cada vez que el número 3 aparece en la lista
print(sum(1 for x in lista_nueva if x == 3))

# Buscamos el índice usando la función enumerate()
print([i for i, x in enumerate(lista_nueva) if x == 2][0])

del lista_nueva[1] # Elimina el elemento en el índice 1 (que es el número 2)

print(lista_nueva)
```

```python
tupla = "¿La tierra es plana?", True, False # Los paréntesis son opcionales
print(tupla)

p0, p1, p2 = tupla # Desempaquetado de tupla
print(p0)
print(p1)
print(p2)

# Alternativa a .count()
print(sum(1 for x in tupla if x is True))

# Alternativa a .index()
print([i for i, x in enumerate(tupla) if x is False][0])

print(int(1)) # Mismo resultado que print((1))
print(tuple([1])) # Mismo resultado que print((1,))
```

```python
print(set())

# Creación directa con llaves (los duplicados se eliminan solos)
print({5, 2, 1, 1.5})
print({5, 2, 1, 1.5})
print(set("52511.5"))

conjunto = {2, 3, 4}
conjunto_2 = {5, 3, 6}
conjunto_3 = {4, 2}

print(conjunto)
print(conjunto_2)
print(conjunto_3)

conjunto.update([1]) # Alternativa a .add()
print(conjunto)

conjunto.discard(1) # Hace lo mismo que .remove() pero no da error si no existe
print(conjunto)

print(conjunto & conjunto_2) # Operador '&' equivale a .intersection()
print(conjunto_2 <= conjunto) # Operador '<=' equivale a .issubset()
print(conjunto_3 <= conjunto)
```

```python
diccionario = {1: "Uno", 2: "Dos"}
diccionario.update({3: "Tres"}) # Alternativa a diccionario[3] = ...
print(diccionario)

dict_lista_tuplas = dict([(1, "Uno"), (2, "Dos"), (3, "Tres")])
print(dict_lista_tuplas)

dict_lista_string = dict(Uno = 1, Dos = 2, Tres = 3)
print(dict_lista_string)

dict_tipos = {1: "integer",
              2.2: "float",
              "texto": "string",
              (1, 2): "tupla"}

print(dict_tipos)

dict_repeticion = {1: "Primero", 1: "Último"}
print(dict_repeticion)

print(diccionario,
      diccionario.keys(),
      diccionario.values(),
      diccionario.items())

claves = diccionario.values()
print(claves)
diccionario.update({1: "One"})
print(claves)
del diccionario[2] # Alternativa a .pop(2)
print(diccionario)
```
