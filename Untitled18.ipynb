```python
import io
f_capitulos1 = io.BytesIO(uploaded['f_capitulos.csv'])
f_capitulos1 = pd.read_csv(f_capitulos1, sep = ",", encoding="latin-1")
f_capitulos1.head(4)
```

```python
tabla_alcohol = pd.DataFrame({
    'edad': f_capitulos1['F_04'],
    'borrachera': f_capitulos1['F_09'],
    'cerveza': f_capitulos1['F_10_A'],
    'recordar': f_capitulos1['F_19'],
    "preocupacion": f_capitulos1["F_21"]
    })
tabla_alcohol.head(5)
```

```python
tabla_alcohol['cerveza'] = tabla_alcohol['cerveza'].fillna(0)
tabla_alcohol.head(5)
```

```python
tabla_alcohol['borrachera'] = tabla_alcohol['borrachera'].fillna(0)
tabla_alcohol.head(5)
```

```python
tabla_alcohol['recordar'] = tabla_alcohol['recordar'].fillna(0) # 0 lo tomaremos para quienes no consumen alcohol
tabla_alcohol.head(5)
```

```python

tabla_alcohol['preocupacion'] = tabla_alcohol['borrachera'].fillna(2)
tabla_alcohol.head(5)
```

```python
tabla_alcohol1 = tabla_alcohol.dropna()
tabla_alcohol1.head(5)
```

```python
tabla_alcohol1['cerveza'] = tabla_alcohol1['cerveza'].astype(str)
tabla_alcohol1['recordar'] = tabla_alcohol1['recordar'].astype(str)
tabla_alcohol1['preocupacion'] = tabla_alcohol1['preocupacion'].astype(str)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('f_capitulos.csv')

edad_inicio = df['F_04'].dropna()
edad_inicio = edad_inicio[edad_inicio < 90]

plt.figure(figsize=(10, 6))
sns.histplot(edad_inicio, bins=20, kde=True, color='skyblue', edgecolor='black')

plt.title('Distribución de la Edad de Inicio de Consumo de Alcohol (Variable F_04)', fontsize=14)
plt.xlabel('Edad (Años)', fontsize=12)
plt.ylabel('Frecuencia (Número de personas)', fontsize=12)
plt.grid(axis='y', alpha=0.3)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="white", palette="muted")
plt.rcParams['font.family'] = 'sans-serif'

df_cerveza = f_capitulos1['F_10_A'].dropna().astype(str).value_counts().reset_index()
df_cerveza.columns = ['Respuesta', 'Total']

df_cerveza = df_cerveza.sort_values('Total', ascending=False)

plt.figure(figsize=(10, 6))
ax = sns.barplot(
    data=df_cerveza,
    x='Respuesta',
    y='Total',
    palette='YlOrBr',
    edgecolor='0.2'
)

plt.title('Consumo de Cerveza (Variable F_10_A)', fontsize=18, fontweight='bold', pad=20, color='#333333')
plt.xlabel('Categoría de Respuesta', fontsize=12, fontweight='semibold')
plt.ylabel('Cantidad de Registros', fontsize=12, fontweight='semibold')

sns.despine(left=True, bottom=False)

for p in ax.patches:
    ax.annotate(f'{int(p.get_height()):,}',
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha = 'center', va = 'center',
                xytext = (0, 9),
                textcoords = 'offset points',
                fontsize=11, fontweight='bold', color='#555555')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# --- 1. Preparación de datos ---
# Mapeo opcional (si tus datos son 1 y 2, cámbialos por nombres reales)
# f_capitulos1['F_10_A'] = f_capitulos1['F_10_A'].replace({1: 'Consume', 2: 'No Consume'})

counts = f_capitulos1['F_10_A'].value_counts().reset_index()
counts.columns = ['Categoria', 'Valor']
counts = counts.sort_values('Valor', ascending=False)

# --- 2. Configuración de Estilo ---
plt.figure(figsize=(11, 7), facecolor='#FDFCF9') # Fondo crema muy tenue
ax = plt.gca()
ax.set_facecolor('#FDFCF9')

# Paleta de colores inspirada en tonos de cerveza (dorado, ámbar, tostado)
colores = sns.color_palette("YlOrBr_r", len(counts))

# --- 3. Gráfico de Barras ---
bars = ax.bar(counts['Categoria'], counts['Valor'], color=colores,
              edgecolor='white', linewidth=2, width=0.7)

# --- 4. Detalles Visuales "Premium" ---
# Añadir etiquetas de datos con formato de miles
for bar in bars:
    height = bar.get_height()
    ax.annotate(f'{int(height):,}',
                xy=(bar.get_x() + bar.get_width() / 2, height),
                xytext=(0, 8),  # 8 puntos de offset vertical
                textcoords="offset points",
                ha='center', va='bottom',
                fontsize=13, fontweight='bold', color='#4A3620')

# --- 5. Títulos y Etiquetas ---
plt.title('Consumo de Cerveza en el Hogar (Variable F_10_A)',
          fontsize=22, fontweight='bold', color='#2C1E11', pad=30, loc='left')

plt.xlabel('Frecuencia / Respuesta', fontsize=14, color='#5D4037', labelpad=15)
plt.ylabel('Total de Encuestados', fontsize=14, color='#5D4037', labelpad=15)

# Limpiar el diseño (quitar bordes innecesarios)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_color('#D7CCC8')
ax.spines['bottom'].set_color('#D7CCC8')

# Ajustar las rejillas (solo horizontales para no ensuciar)
ax.yaxis.grid(True, linestyle='--', alpha=0.3, color='#8D6E63')
ax.set_axisbelow(True)

plt.tight_layout()
plt.show()
```
