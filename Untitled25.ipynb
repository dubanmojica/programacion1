python permite cargr y crear objetos de archivos dados, y tambien oermite crear archivos y editar archivos dados


```python
# crear y escribir un archivo de texto
with open("archivo.txt", "w") as f:
    f.write("Hola, este es un archivo de texto.\n")
    f.write("segunda línea.\n")
```

con el argumento "w" se crea el archivo si no existe. tambien sobrescribe el archivo si ya existe, borrando todo su contenido previo

```python
with open("archivo.txt", "r") as f:
    print(f.read())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# The 'Ciudad' column is at index 2 in the DataFrame 'data1'
ciudad_counts = data1[2].value_counts()

plt.figure(figsize=(10, 6))
sns.barplot(x=ciudad_counts.index, y=ciudad_counts.values)
plt.title('Distribución de Ciudades')
plt.xlabel('Ciudad')
plt.ylabel('Cantidad')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
