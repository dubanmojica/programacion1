```python
def validar_numero(mensaje):
    while True:
        try:
            valor = float(input(mensaje))
            return valor
        except ValueError:
            print("Error: Debe ingresar un número válido.")

def ingresar_matriz():
    print("\n--- Configuración de Matriz ---")
    filas = int(validar_numero("Ingrese el número de filas: "))
    columnas = int(validar_numero("Ingrese el número de columnas: "))

    matriz = []
    for i in range(filas):
        fila = []
        for j in range(columnas):
            elemento = validar_numero(f"Elemento [{i+1}][{j+1}]: ")
            fila.append(elemento)
        matriz.append(fila)
    return matriz

def mostrar_matriz(matriz):
    if matriz is None:
        return
    for fila in matriz:
        print("[", end=" ")
        for elemento in fila:
            print(f"{elemento:8.2f}", end=" ")
        print("]")
```

```python
def sumar_matrices(A, B):
    if len(A) != len(B) or len(A[0]) != len(B[0]):
        print("Error: Las matrices deben tener las mismas dimensiones.")
        return None
    return [[A[i][j] + B[i][j] for j in range(len(A[0]))] for i in range(len(A))]

def multiplicar_matrices(A, B):
    if len(A[0]) != len(B):
        print("Error: Columnas de A deben coincidir con filas de B.")
        return None
    resultado = [[0 for _ in range(len(B[0]))] for _ in range(len(A))]
    for i in range(len(A)):
        for j in range(len(B[0])):
            for k in range(len(B)):
                resultado[i][j] += A[i][k] * B[k][j]
    return resultado

def hadamard(A, B):
    if len(A) != len(B) or len(A[0]) != len(B[0]):
        print("Error: Las dimensiones deben ser iguales para el producto Hadamard.")
        return None
    return [[A[i][j] * B[i][j] for j in range(len(A[0]))] for i in range(len(A))]

def kronecker(A, B):
    filas_A, col_A = len(A), len(A[0])
    filas_B, col_B = len(B), len(B[0])
    resultado = [[0 for _ in range(col_A * col_B)] for _ in range(filas_A * filas_B)]

    for i in range(filas_A):
        for j in range(col_A):
            for k in range(filas_B):
                for l in range(col_B):
                    resultado[i * filas_B + k][j * col_B + l] = A[i][j] * B[k][l]
    return resultado
```

```python
def mostrar_menu():
    print("\n" + "-"*10 + " MENU DE OPERACIONES CON MATRICES " + "-"*10)
    print("1. Suma de matrices")
    print("2. Multiplicación de matrices")
    print("3. Producto de Hadamard")
    print("4. Producto de Kronecker")
    print("5. Salir")

    while True:
        try:
            opcion = int(input("\nSeleccione una opción: "))
            if opcion in [1, 2, 3, 4, 5]:
                return opcion
            else:
                print("Error: Opción no válida.")
        except ValueError:
            print("Error: Debe ingresar un número entero.")
```

```python
def ejecutar():
    while True:
        opcion = mostrar_menu()

        if opcion == 5:
            print("Saliendo del programa...")
            break

        print("\n--- Ingrese la Matriz A ---")
        A = ingresar_matriz()
        print("\n--- Ingrese la Matriz B ---")
        B = ingresar_matriz()
        resultado = None
        if opcion == 1:
            resultado = sumar_matrices(A, B)
        elif opcion == 2:
            resultado = multiplicar_matrices(A, B)
        elif opcion == 3:
            resultado = hadamard(A, B)
        elif opcion == 4:
            resultado = kronecker(A, B)
        if resultado:
            print("\nResultado de la operación:")
            mostrar_matriz(resultado)
if __name__ == "__main__":
    ejecutar()
```
