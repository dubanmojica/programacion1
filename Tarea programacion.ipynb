```python
lista = ["Daniel", 17, 3208908522, "Pasto"]
print("lista", lista)
tupla = tuple(lista)
print("lista a tupla:",tupla)
conjunto = set(lista)
print("lista a conjunto:",conjunto)
diccionario = dict(enumerate(lista))
print("lista a diccionario:",diccionario)

```

```python
tupla = ("Daniel", 17, 3208908522, "Pasto")
print("\nTUPLA:", tupla)
lista = list(tupla)
print("tupla a lista:", lista)
conjunto = set(tupla)
print("tupla a conjunto:", conjunto)
diccionario = dict(enumerate(tupla))
print("tupla a diccionario:", diccionario)
```

```python

diccionario = {
    "Nombre": "Daniel",
    "Edad": 17,
    "Celular": 3208908522,
    "Ciudad": "Pasto"
}

print("\nDICCIONARIO:", diccionario)

# Diccionario → Lista
lista = list(diccionario.items())
print("Diccionario a Lista:", lista)

# Diccionario → Tupla
tupla = tuple(diccionario.items())
print("Diccionario a Tupla:", tupla)

# Diccionario → Conjunto
conjunto = set(diccionario.items())
print("Diccionario a Conjunto:", conjunto)
```
