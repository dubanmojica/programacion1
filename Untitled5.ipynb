```python
diccionario = {1: "Uno", 2: "Dos"}
diccionario.update({3: "Tres"}) # cambie el (diccionario[3] = "Tres") por un metodo que los agrega de una vez y no uno por uno
print(diccionario)

dict_lista_tuplas = dict([(1, "Uno"), (2, "Dos"), (3, "Tres")])
print(dict_lista_tuplas)

dict_lista_string = dict(Uno = 1, Dos = 2, Tres = 3)
print(dict_lista_string)

dict_tipos = {1: "integer",
              2.2: "float",
              "texto": "string",
              (1, 2): "tupla"}

print(dict_tipos)

dict_repeticion = {1: "Primero", 1: "Último"}
print(dict_repeticion)

print(diccionario,
      diccionario.keys(), # keys devuelve la lista con las claves
      diccionario.values(), # values devuelve una lista con los valores
      diccionario.items()) # devuelve una lista de tuplas con los pares de datos keys y values

claves = diccionario.values()
print(claves)
diccionario[1] = "One"
for valor in diccionario.values(): #cambiamos (print(claves))para que los codigos sea mas legibles al imprimirlos
    print(valor)
diccionario.pop(2) #para eliminar una clave se utiliza pop
print(diccionario)
```

```python
print(0, "->", bool(0))
print(0.0, "->", bool(0.0))
print("", "->", bool(""))
print(list(), "->", bool(list()))
print(tuple(), "->", bool(tuple()))
print(set(), "->", bool(set()))
print(dict(), "->", bool(dict()))
print(None, "->", bool(None))
```

se hizo ese cambio para poner todos los objetos en una lista y recorrerlos. Esto hace que el código sea mucho más fácil de mantener para agregar más ejemplos en el futuro.

```python
elementos = [0, 0.0, "", list(), tuple(), set(), dict(), None]

for elemento in elementos:
    print(f"{elemento} -> {bool(elemento)}")
```

```python
def calificar_nota(nota):
    if not (0 <= nota <= 10):
        return "Nota fuera de rango"
    elif nota < 5:
        return "Estás suspenso, F"
    elif nota < 7:
        return "Aprobado"
    elif nota < 9:
        return "Notable"
    else:
        return "Sobresaliente"
try:
    valor = float(input("Introduzca su nota: "))
    print(calificar_nota(valor))
except ValueError:
    print("Por favor, introduce un número válido.")
```
